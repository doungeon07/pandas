# Resampling

```python
df.Close.resample('M').mean()
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot()
```

```python

%matplotlib inline
df.Close.resample('W').mean().plot()
```

```python
%matplotlib inline # quarterly
df.Close.resample('Q').mean().plot()
```

```python
%matplotlib inline 
df.Close.resample('Q').mean().plot(kind='bar')
```
